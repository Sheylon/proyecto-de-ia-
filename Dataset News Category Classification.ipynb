En esta etapa se importa la librería NLTK y se descargan algunos recursos necesarios para el procesamiento de lenguaje natural, como palabras vacías, lematización y tokenización.

````python
import nltk

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
````

````python
stop_words = set(stopwords.words('english'))

lemmatizer = WordNetLemmatizer()
````

````python
import re

def procesar_texto(texto):

    # convertir a string
    texto = str(texto)

    # minúsculas
    texto = texto.lower()

    # eliminar urls
    texto = re.sub(r'http\S+|www\S+', '', texto)

    # eliminar caracteres especiales y números
    texto = re.sub(r'[^a-zA-Z\s]', '', texto)

    # tokenizar
    palabras = word_tokenize(texto)

    # eliminar stopwords y lematizar
    palabras_limpias = [
        lemmatizer.lemmatize(palabra)
        for palabra in palabras
        if palabra not in stop_words
        and len(palabra) > 2
    ]

    # unir texto nuevamente
    return ' '.join(palabras_limpias)
````

````python
noticias['clean_headline'] = noticias['clean_headline'].apply(procesar_texto)

noticias['clean_description'] = (noticias['clean_description'].fillna('').apply(procesar_texto))
````

## Análisis de similitud textual entre categorías

Con el objetivo de identificar posibles categorías relacionadas entre sí, se realizó un análisis de similitud textual entre las categorías del conjunto de datos.

Para esto, se combinaron los textos de cada noticia usando el titular limpio (`clean_headline`) y la descripción limpia (`clean_description`). Posteriormente, se agruparon todos los textos por categoría, de manera que cada categoría quedara representada por un único conjunto de palabras.

Luego se aplicó la técnica TF-IDF, la cual permite representar cada categoría como un vector numérico basado en la importancia de sus palabras. Finalmente, se calculó la similitud del coseno entre estos vectores, con el fin de medir qué tan parecido es el vocabulario entre las categorías.

Los valores de similitud cercanos a 1 indican que dos categorías comparten un vocabulario muy similar, mientras que valores cercanos a 0 indican menor relación textual.

Este análisis permite identificar categorías que podrían estar relacionadas temáticamente y que, en algunos casos, podrían agruparse para reducir la complejidad del problema de clasificación.

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear texto combinado
noticias['texto_modelo'] = (
    noticias['clean_headline'].fillna('').astype(str) + ' ' +
    noticias['clean_description'].fillna('').astype(str)
)

# Unir todos los textos por categoría
textos_por_categoria = noticias.groupby('category')['texto_modelo'].apply(
    lambda textos: ' '.join(textos.astype(str))
)

# Vectorizar cada categoría usando TF-IDF
vectorizador = TfidfVectorizer(
    stop_words='english',
    max_features=15000,
    ngram_range=(1, 2)
)

tfidf_categorias = vectorizador.fit_transform(textos_por_categoria)

# Calcular similitud entre categorías
matriz_similitud = cosine_similarity(tfidf_categorias)

# Convertir a DataFrame para visualizar mejor
similitud_categorias = pd.DataFrame(
    matriz_similitud,
    index=textos_por_categoria.index,
    columns=textos_por_categoria.index
)

# Mostrar matriz de similitud
similitud_categorias
````

````python
plt.figure(figsize=(18, 14))

sns.heatmap(
    similitud_categorias,
    cmap='Blues',
    linewidths=0.5
)

plt.title('Similitud textual entre categorías usando TF-IDF y coseno')
plt.xlabel('Categoría')
plt.ylabel('Categoría')
plt.show()
````

## Categorías con mayor similitud textual

Para identificar qué categorías podrían estar relacionadas, se transformó la matriz de similitud en una tabla de pares de categorías. Antes de hacer esta transformación, se eliminó la diagonal principal, ya que representa la comparación de cada categoría consigo misma.

También se ajustaron los nombres del índice y de las columnas para evitar conflictos al convertir la matriz en una tabla. Finalmente, se eliminaron los pares repetidos y se ordenaron las combinaciones de mayor a menor similitud.

Esta tabla permite observar qué categorías comparten mayor vocabulario y podrían considerarse candidatas para una posible agrupación temática.

````python
import numpy as np
import pandas as pd

# Copiar matriz de similitud
pares_similares = similitud_categorias.copy()

# Eliminar diagonal principal para no comparar una categoría consigo misma
np.fill_diagonal(pares_similares.values, np.nan)

# Quitar el nombre del índice y de las columnas para evitar el error
pares_similares.index.name = 'categoria_1'
pares_similares.columns.name = 'categoria_2'

# Pasar de matriz a formato tabla
pares_similares = pares_similares.stack().reset_index()

# Renombrar columnas
pares_similares.columns = ['categoria_1', 'categoria_2', 'similitud']

# Crear una columna para evitar pares repetidos
pares_similares['par_ordenado'] = pares_similares.apply(
    lambda x: tuple(sorted([x['categoria_1'], x['categoria_2']])),
    axis=1
)

# Eliminar pares repetidos
pares_similares = pares_similares.drop_duplicates('par_ordenado')

# Eliminar columna auxiliar
pares_similares = pares_similares.drop(columns='par_ordenado')

# Ordenar de mayor a menor similitud
pares_similares = pares_similares.sort_values(by='similitud', ascending=False)

# Mostrar los 20 pares más similares
pares_similares.head(50)
````

Categorías con mayor similitud textual

La tabla anterior muestra los pares de categorías con mayor similitud textual según el vocabulario presente en sus titulares y descripciones.

Estos resultados pueden servir como una guía para identificar categorías que podrían agruparse por afinidad temática. Por ejemplo, si dos categorías comparten muchas palabras frecuentes, es posible que traten temas similares o que exista solapamiento entre sus contenidos.

**Aplicacion de chi-Squared**

Primero, se importan las librerías necesarias para convertir texto en datos numéricos, codificar categorías y aplicar el algoritmo de selección de características Chi-Cuadrado.

Luego, se utiliza TF-IDF (Term Frequency - Inverse Document Frequency) para transformar el texto de las noticias en una matriz numérica. Este método asigna mayor importancia a las palabras más relevantes dentro de cada noticia y reduce la importancia de palabras demasiado comunes. Además, se limita el vocabulario a un máximo de 30.000 palabras y se eliminan palabras vacías del idioma inglés.

Posteriormente, las categorías de las noticias se convierten en valores numéricos mediante LabelEncoder, permitiendo que el modelo pueda trabajar con ellas.

Después, se aplica el algoritmo Chi-Cuadrado mediante SelectKBest para identificar las palabras más importantes en la clasificación de las noticias. En este caso, se seleccionan las 12.000 características más relevantes según su relación con las categorías del dataset.

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, chi2

# 1. Vectorizar los datos de texto usando TF-IDF
tfidf_vectorizer = TfidfVectorizer(
    max_features=30000,
    stop_words='english',
    min_df=5,
    max_df=0.8,
    sublinear_tf=True
)  # Puedes ajustar max_features

X = tfidf_vectorizer.fit_transform(
    noticias['texto_modelo'].fillna('')
)

# 2. Codificar la variable objetivo (categoría) en etiquetas numéricas
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(noticias['category'])

# 3. Aplicar selección de características con Chi-Cuadrado
# Seleccionar las mejores K características
k_best_features = SelectKBest(chi2, k=15000)  # Ajusta k según necesites

X_new = k_best_features.fit_transform(X, y)

# Obtener los nombres de las características seleccionadas
selected_feature_indices = k_best_features.get_support(indices=True)

selected_feature_names = (
    tfidf_vectorizer.get_feature_names_out()
    [selected_feature_indices]
)

print(f"Número original de características: {X.shape[1]}")

print(f"Número de características después de Chi-Cuadrado: {X_new.shape[1]}")

print("Principales características seleccionadas (primeras 20):")

print(selected_feature_names[:20])

# También puedes obtener los puntajes de las características
chi2_scores = k_best_features.scores_[selected_feature_indices]

feature_scores = pd.DataFrame({
    'feature': selected_feature_names,
    'score': chi2_scores
})

feature_scores = feature_scores.sort_values(
    by='score',
    ascending=False
)

print("\nTop 20 características según el puntaje Chi-Cuadrado:")

print(feature_scores.head(20))
````

Finalmente, el código obtiene y muestra:

- El número original de palabras del vocabulario.
- La cantidad de palabras seleccionadas después del filtrado.
- Las principales palabras seleccionadas.
- Las 20 palabras con mayor puntuación Chi-Cuadrado, es decir, las más importantes para distinguir entre categorías de noticias.

Este proceso permite reducir ruido, mejorar la calidad de los datos y optimizar el rendimiento de los modelos de Machine Learning posteriores.
